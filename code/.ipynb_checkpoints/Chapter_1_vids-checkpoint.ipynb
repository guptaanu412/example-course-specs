#### Learn to fit k-nn: Cancer Data

```python
from sklearn.neighbors import KNeighborsClassifier 
bc_knn = KNeighborsClassifier(n_neighbors=6)
bc_knn.fit(bc_data['data'], bc_data['target'])
```

#### Predict on new data: Cancer Data

```python
X_new = np.random.rand(1,30)*200
prediction = bc_knn.predict(X_new)
print("Prediction: {}".format(prediction)) 
print("Predicted target name: {}".format(
    bc_data['target_names'][prediction]))
```

```python
X_new = np.reshape(bc_data['data'][10,:], (1, 30)) + np.random.rand(1,30)
prediction = bc_knn.predict(X_new)
print("Prediction: {}".format(prediction)) 
print("Predicted target name: {}".format(
    bc_data['target_names'][prediction]))
```

## Exercise 8 (Video): Overfitting, Underfitting and Train/Test Split
Demonstrate (using cancer data) the need to split data into training and testing sets.

Discuss bias-variance tradeoff.

#### Cancer Data: Train/Test Split and Fit

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=42)

bc_knn = KNeighborsClassifier(n_neighbors=17)
bc_knn.fit(X_train, y_train)
```

#### Cancer Data: Train/Test Split + Fit/Predict

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

bc_X_train, bc_X_test, bc_y_train, bc_y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=21)
        
bc_knn = KNeighborsClassifier(n_neighbors=17)
bc_knn.fit(bc_X_train, bc_y_train)
bc_y_pred = bc_knn.predict(bc_X_test)
print("Test set predictions:\n {}".format(bc_y_pred))
print("Test set score: {:.2f}".format(np.mean(bc_y_pred == bc_y_test)))
```

#### Cancer Data: Effect of varying k in k-nn

```python
neighbors = np.arange(1, 26)
train_accuracy = np.empty(len(neighbors))
test_accuracy = np.empty(len(neighbors))

for i, k in enumerate(neighbors):
    # Setup a KNN Classifier
    clf = KNeighborsClassifier(n_neighbors = k)

    # Fit the classifier to the training data
    clf.fit(bc_X_train, bc_y_train)

    #Compute accuracy on the training set
    train_accuracy[i] = accuracy_score(bc_y_train, clf.predict(bc_X_train))

    #Compute accuracy on the testing set
    test_accuracy[i] = accuracy_score(bc_y_test, clf.predict(bc_X_test))
    
plt.title('KNN: Varying Number of Neighbors')
plt.plot(neighbors, test_accuracy, label = 'Testing Accuracy')
plt.plot(neighbors, train_accuracy, label = 'Training Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```
