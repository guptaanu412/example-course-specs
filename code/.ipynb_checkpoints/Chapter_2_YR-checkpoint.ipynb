### Exercise 3: Importing Data for Supervised Learning

#### Boston Data

```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

boston = pd.read_csv('../datasets/boston.csv', header = None,
                 delim_whitespace = True, names = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE',
                                                   'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'MEDV'])

boston.head()
```

```python
bos_y = boston['MEDV']
bos_X = boston.drop('MEDV', axis=1)
```

```python
#Draw attention to MEDV's high correlation with RM
_ = pd.scatter_matrix(boston, c=bos_y, figsize=(25,25))
```

```python
print(np.shape(boston))
print(boston.columns)
```

### Exercise 4: The Basics of Linear Regression (Video)

```python
##Plot Value of house as a function of number of rooms
bos_yb = bos_y.reshape(-1, 1)
bos_Xb = bos_X.iloc[:,5].reshape(-1, 1)
# Plot data
_ = plt.scatter(bos_Xb, bos_yb)
_ = plt.ylabel('value of house /1000 ($)')
_ = plt.xlabel('number of rooms')
```

Linear model with one predictor variable:

```python
from sklearn import linear_model
# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(bos_Xb, bos_yb)
# Plot outputs
plt.scatter(bos_Xb, bos_yb,  color='black')
plt.plot(bos_Xb, regr.predict(bos_Xb), color='blue',
         linewidth=3)
plt.show()
```

```python
print(regr.score(bos_Xb, bos_yb))

#Predict with 1 variable
print(regr.predict(7))
```

```python
from sklearn.metrics import mean_squared_error

#Use all predictor variables
bos_Xb = bos_X

# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit(bos_Xb, bos_yb)
bos_mod = regr_all.predict(bos_Xb)

print(regr_all.score(bos_Xb, bos_yb))
rmse = np.sqrt(mean_squared_error(bos_yb, bos_mod))
print(rmse)
```

```python
from sklearn.model_selection import train_test_split

bos_X_train, bos_X_test, bos_y_train, bos_y_test = train_test_split(bos_X, bos_yb, test_size = 0.3, random_state=42)
```

```python
# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit(bos_X_train, bos_y_train)
bos_mod = regr_all.predict(bos_X_test)

regr_all.score(bos_X_test, bos_y_test)
```

```python
rmse = np.sqrt(mean_squared_error(bos_y_test, bos_mod))
print(rmse)
```

### Exercise 5: Linear Regression: Fit and Predict

#### Gapminder Data

```python
gm = pd.read_csv('../datasets/gm_2008.csv')
```

```python
gm_Xs = gm.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).values
gm_y = gm.dropna(how='any').child_mortality.values
gm_y = np.asarray(gm.dropna(how='any').child_mortality)
gm_names = gm.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).columns
np.shape(gm_Xs), np.shape(gm_y)
```

```python
##Plot GDP as a function of fertility

_ = plt.scatter(gm_Xs[:,1],gm_y)
_ = plt.ylabel('GDP')
_ = plt.xlabel('Fertility')
```

```python
#Linear Regression with one predictor variable

gm_y = gm_y.reshape(-1, 1)
gm_Xb = gm_Xs[:,1].reshape(-1, 1)

# Create linear regression object
regr = linear_model.LinearRegression()
# Train the model using the training sets
regr.fit(gm_Xb, gm_y)
# Plot outputs
plt.scatter(gm_Xb, gm_y,  color='black')
plt.plot(gm_Xb, regr.predict(gm_Xb), color='blue',
         linewidth=3)
plt.show()
```

```python
print(regr.score(gm_Xb, gm_y))

#Predict with 1 variable
print(regr.predict(3))
```

```python
# All Variables
# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit(gm_Xs, gm_y)
gm_mod = regr_all.predict(gm_Xs)

print(regr_all.score(gm_Xs, gm_y))
rmse = np.sqrt(mean_squared_error(gm_y, gm_mod))
print(rmse)
```

### Exercise 6: Train/Test Split for Regression

```python
gm_X_train, gm_X_test, gm_y_train, gm_y_test = train_test_split(gm_Xs, gm_y, test_size = 0.3, random_state=42)

# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit(gm_X_train, gm_y_train)
gm_mod = regr_all.predict(gm_X_test)

print(regr_all.score(gm_X_test, gm_y_test))
rmse = np.sqrt(mean_squared_error(gm_y_test, gm_mod))
print(rmse)
```

### Exercise 7: Cross-Validation (Video)

#### Boston Data

```python
from sklearn.model_selection import cross_val_predict
from sklearn.model_selection import cross_val_score

regr = linear_model.LinearRegression()

# cross_val_predict returns an array of the same size as `y` where each entry
# is a prediction obtained by cross validation:
predicted = cross_val_predict(regr, bos_X, bos_y, cv=5)

plt.scatter(bos_y, predicted, color='black')
plt.plot([bos_y.min(), bos_y.max()], [bos_y.min(), bos_y.max()], color = 'blue', linewidth=3)
plt.ylabel('Predicted')
plt.xlabel('Measured')
plt.show();
```

```python
model = linear_model.LinearRegression()

cross_results = cross_val_score(model, bos_X, bos_y, cv = 5)
for i in cross_results:
    print(i)
print("Average 5-Fold CV Score:" + str(np.mean(cross_results)))
```

```python
elastic_net_coef
```

```python
##code closely modeled on this: http://scikit-learn.org/stable/auto_examples/exercises/plot_cv_diabetes.html#sphx-glr-auto-examples-exercises-plot-cv-diabetes-py
## so will need to edit
from sklearn.model_selection import cross_val_score
alphas = np.logspace(-4, 1, 30)

lasso = Lasso()

scores = list()
scores_std = list()

n_folds = 10

for alpha in alphas:
    lasso.alpha = alpha
    this_scores = cross_val_score(lasso, bos_X, bos_yb, cv=n_folds, n_jobs=1)
    scores.append(np.mean(this_scores))
    scores_std.append(np.std(this_scores))

scores, scores_std = np.array(scores), np.array(scores_std)
plt.figure().set_size_inches(8, 6)
plt.semilogx(alphas, scores)

# plot error lines showing +/- std. errors of the scores
std_error = scores_std / np.sqrt(n_folds)

plt.semilogx(alphas, scores + std_error, 'b--')
plt.semilogx(alphas, scores - std_error, 'b--')

# alpha=0.2 controls the translucency of the fill color
plt.fill_between(alphas, scores + std_error, scores - std_error, alpha=0.2)

plt.ylabel('CV score +/- std error')
plt.xlabel('alpha')
plt.axhline(np.max(scores), linestyle='--', color='.5')
plt.xlim([alphas[0], alphas[-1]])
```
