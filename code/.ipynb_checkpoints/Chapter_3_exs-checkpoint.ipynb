```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn import datasets
```

## Exercise 2: Metrics for Classification

          Compute precision, recall, and f-1 score on voting dataset. Use
          confusion_matrix and classification_report from sklearn.metrics,
          DecisionTreeClassifier() and voting data.

### Voting Data

```python
df = pd.read_csv('../datasets/house-votes-84.csv',
                header=None, names = ['infants', 'water', 'budget', 'physician', 'salvador', 'religious',
                                     'satellite', 'aid', 'missile', 'immigration', 'synfuels', 'education',
                                     'superfund', 'crime', 'duty_free_exports', 'eaa_rsa'])

df = df.reset_index()
df.rename(columns = {'index': 'party'}, inplace = True)

df[df == 'y'] = 1
df[df == 'n'] = 0
df[df == '?'] = np.nan
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.fillna(x.mean()))
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

y = df['party']
X = df.drop('party', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

y_pred = (y_pred == 'democrat').astype(int)
y_test = (y_test == 'democrat').astype(int)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Recall: " + str(recall_score(y_test, y_pred)))
print("Precision: " + str(precision_score(y_test, y_pred)))
print("Recall: " + str(f1_score(y_test, y_pred)))
```

## Exercise 3: Precision-Recall curve

          Explore precision and recall with a precision-recall curve. Use
          precision_recall_curve from sklearn.metrics.

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_pred)

plt.figure(figsize=(6, 6), dpi=None)
plt.plot(recall, precision, label='Decision Tree Classifier')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='best')
plt.show()
```

## Exercise 4: AUC Computation

          Compute AUC and plot ROC curve for voting dataset. Use
          DecisionTreeClassifier() and roc_auc_score().

```python
print("AUC: " + str(roc_auc_score(y_test, y_pred)))
fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label=1)
plt.figure(figsize=(6, 6), dpi=None)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Decision Tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

## Exercise 9: Hyperparameter Tuning with GridSearchCV
### Voting Data

```python
from sklearn.model_selection import GridSearchCV

df = pd.read_csv('../datasets/house-votes-84.csv',
                header=None, names = ['infants', 'water', 'budget', 'physician', 'salvador', 'religious',
                                     'satellite', 'aid', 'missile', 'immigration', 'synfuels', 'education',
                                     'superfund', 'crime', 'duty_free_exports', 'eaa_rsa'])

df = df.reset_index()
df.rename(columns = {'index': 'party'}, inplace = True)

df[df == 'y'] = 1
df[df == 'n'] = 0
df[df == '?'] = np.nan
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.fillna(x.mean()))

y = df['party']
X = df.drop('party', axis=1)

tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

clf = DecisionTreeClassifier()

#Note: You may get 'voting_cv' not defined. If so, run cell once without %timeit
voting_cv = GridSearchCV(clf, tree_param_grid, cv = 5).fit(X, y)

print("Tuned Decision Tree Parameters: " + str(voting_cv.best_params_))
```

## Exercise 13: Hold-out Set in Practice I: Classification

          Using the voting dataset, create a hold-out set, tune the
          hyperparameters of an SVM using GridSearchCV, and assess performance
          against the hold-out set.

```python
df = pd.read_csv('../datasets/house-votes-84.csv',
                header=None, names = ['infants', 'water', 'budget', 'physician', 'salvador', 'religious',
                                     'satellite', 'aid', 'missile', 'immigration', 'synfuels', 'education',
                                     'superfund', 'crime', 'duty_free_exports', 'eaa_rsa'])

df = df.reset_index()
df.rename(columns = {'index': 'party'}, inplace = True)

df[df == 'y'] = 1
df[df == 'n'] = 0
df[df == '?'] = np.nan
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.fillna(x.mean()))

y = df['party']
X = df.drop('party', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)

tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

clf = DecisionTreeClassifier()

voting_cv = RandomizedSearchCV(clf, tree_param_grid, cv = 5).fit(X_train, y_train)


y_pred = voting_cv.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Tuned Decision Tree Parameters: " + str(voting_cv.best_params_))
print("Tuned Decision Tree Accuracy: " + str(acc))
```

## Exercise 14: Hold-out Set in Practice II: Regression

          Using the gapminder dataset, create a hold-out set, tune the
          hyperparameters of an Elastic Net using RandomizedSearchCV, and assess
          performance against the hold-out set.

```python
from sklearn.linear_model import ElasticNet
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

df = pd.read_csv('../datasets/gm_2008.csv')
df.drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).head()
Xs = df.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).values
y = df.dropna(how='any').child_mortality.values
y = np.asarray(df.dropna(how='any').child_mortality)
names = df.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).columns

from sklearn.preprocessing import scale
X = scale(Xs)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)

param_grid = {'alpha': np.logspace(-4, 1, 30)}

elastic_net = ElasticNet()

gm_cv = RandomizedSearchCV(elastic_net, param_grid, cv = 5).fit(X_train, y_train)


y_pred = gm_cv.predict(X_test)
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
print("Tuned ElasticNet Alpha: " + str(gm_cv.best_params_))
print("Tuned ElasticNet R squared: " + str(r2))
print("Tuned ElasticNet MSE: " + str(mse))
```
