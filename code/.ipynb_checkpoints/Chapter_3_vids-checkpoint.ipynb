## Chapter 3: Fine-tuning your model

      Having trained your model, your next task is to evaluate its performance.
      What metrics can you use to gauge how good your model is? So far, you have
      used accuracy for classification and R-squared for regression. In this
      chapter, you will learn about some of the other metrics available in 
      scikit-learn that will allow you to assess your model's performance in a
      more nuanced manner. You will then learn to optimize both your
      classification as well as regression models using hyperparameter tuning.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn import datasets
```

## Exercise 1 (Video): How good is your model?

          Using a Decision Tree and the breast cancer dataset, introduce the 
          confusion matrix, Type I/II errors, AUC/ROC curve, and precision, 
          recall, f1 score. Discuss the pitfalls of accuracy.

### Cancer Data

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


bc_data = datasets.load_breast_cancer()

bc_X_train, bc_X_test, bc_y_train, bc_y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=42)

bc_tree = DecisionTreeClassifier()

bc_tree.fit(bc_X_train, bc_y_train)
bc_y_pred = bc_tree.predict(bc_X_test)
print(confusion_matrix(bc_y_test, bc_y_pred))
print(classification_report(bc_y_test, bc_y_pred))
print("AUC: " + str(roc_auc_score(bc_y_test, bc_y_pred)))
print("Accuracy Score: " + str(accuracy_score(bc_y_test, bc_y_pred)))
print("Recall: " + str(recall_score(bc_y_test, bc_y_pred)))
print("Precision: " + str(precision_score(bc_y_test, bc_y_pred)))
print("Recall: " + str(f1_score(bc_y_test, bc_y_pred)))
```

```python
fpr, tpr, thresholds = roc_curve(bc_y_test, bc_y_pred, pos_label=1)
plt.figure(figsize=(6, 6), dpi=None)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Decision Tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

          plots to assess a model's validity. Introduce MSE/RMSE as other

## Residuals

#### Boston

```python
from sklearn import linear_model

boston = datasets.load_boston()

y = boston['target']
X = boston['data']

yb = y.reshape(-1, 1)
Xb = X

# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit( Xb, yb)
mod = regr_all.predict(Xb)

#plot residuals
_ = plt.plot(range(len(Xb)), mod - yb, marker='o', linestyle='None')
```

```python
from sklearn.linear_model import Lasso
las = Lasso(alpha=0.1).fit(Xb, yb)
model = las.predict(Xb)
plt.plot(model-y, marker = 'o', linestyle='None')
plt.figure()
plt.plot(y, marker = 'o', linestyle='None')
plt.plot(model)
plt.show()
```

## Exercise 8: Hyperparameter Tuning (Video)
### Cancer Data

          Introduce hyperparameter tuning using the breast cancer dataset.
          Demonstrate GridSearchCV and RandomizedSearchCV.

```python
from sklearn.model_selection import GridSearchCV

bc_data = datasets.load_breast_cancer()



tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

bc_tree = DecisionTreeClassifier()

#Note: You may get 'tree_cv' not defined. If so, run cell once without %timeit
tree_cv = GridSearchCV(bc_tree, tree_param_grid, cv = 5).fit(bc_data['data'], bc_data['target'])

print("Tuned Decision Tree Parameters: " + str(tree_cv.best_params_))
```

```python
from sklearn.model_selection import RandomizedSearchCV

bc_data = datasets.load_breast_cancer()

tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

bc_tree = DecisionTreeClassifier()

%timeit tree_cv = RandomizedSearchCV(bc_tree, tree_param_grid, cv = 5).fit(bc_data['data'], bc_data['target'])

print("Tuned Decision Tree Parameters: " + str(tree_cv.best_params_))
```

## Exercise 11 (Video): Hold-out Set for Final Evaluation

          Introduce the hold-out set as a way to assess the performance of a
          fully specified classifier or regressor with optimized parameters.
          Revisit the ideas of overfitting and train/test split.

```python
bc_data = datasets.load_breast_cancer()

bc_X_train, bc_X_test, bc_y_train, bc_y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=42)

tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

bc_tree = DecisionTreeClassifier()

tree_cv = GridSearchCV(bc_tree, tree_param_grid, cv = 5).fit(bc_X_train, bc_y_train)

#Fit the fully specified classifier with optimized params to new data
bc_pred = tree_cv.predict(bc_X_test)

bc_acc = accuracy_score(bc_y_test, bc_pred)

print("Tuned Decision Tree Parameters: " + str(tree_cv.best_params_))
print("Tuned Decision Tree Accuracy: " + str(bc_acc))
```
