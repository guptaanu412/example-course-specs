```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn import datasets
```

## Exercise 1 (Video): How good is your model?

### Cancer Data

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier


bc_data = datasets.load_breast_cancer()

bc_X_train, bc_X_test, bc_y_train, bc_y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=42)

bc_tree = DecisionTreeClassifier()

bc_tree.fit(bc_X_train, bc_y_train)
bc_y_pred = bc_tree.predict(bc_X_test)
print(confusion_matrix(bc_y_test, bc_y_pred))
print(classification_report(bc_y_test, bc_y_pred))
print("AUC: " + str(roc_auc_score(bc_y_test, bc_y_pred)))
print("Accuracy Score: " + str(accuracy_score(bc_y_test, bc_y_pred)))
print("Recall: " + str(recall_score(bc_y_test, bc_y_pred)))
print("Precision: " + str(precision_score(bc_y_test, bc_y_pred)))
print("Recall: " + str(f1_score(bc_y_test, bc_y_pred)))
```

```python
fpr, tpr, thresholds = roc_curve(bc_y_test, bc_y_pred, pos_label=1)
plt.figure(figsize=(6, 6), dpi=None)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Decision Tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

## Exercise 2: Metrics for Classification

### Voting Data

```python
df = pd.read_csv('../datasets/house-votes-84.csv',
                header=None, names = ['infants', 'water', 'budget', 'physician', 'salvador', 'religious',
                                     'satellite', 'aid', 'missile', 'immigration', 'synfuels', 'education',
                                     'superfund', 'crime', 'duty_free_exports', 'eaa_rsa'])

df = df.reset_index()
df.rename(columns = {'index': 'party'}, inplace = True)

df[df == 'y'] = 1
df[df == 'n'] = 0
df[df == '?'] = np.nan
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.fillna(x.mean()))
```

```python
from sklearn.model_selection import train_test_split

y = df['party']
X = df.drop('party', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

y_pred = (y_pred == 'democrat').astype(int)
y_test = (y_test == 'democrat').astype(int)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print("Recall: " + str(recall_score(y_test, y_pred)))
print("Precision: " + str(precision_score(y_test, y_pred)))
print("Recall: " + str(f1_score(y_test, y_pred)))
```

## Exercise 3: AUC Computation

```python
print("AUC: " + str(roc_auc_score(y_test, y_pred)))
fpr, tpr, thresholds = roc_curve(y_test, y_pred, pos_label=1)
plt.figure(figsize=(6, 6), dpi=None)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Decision Tree Classifier')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()
```

## Exercise 5: Hyperparameter Tuning (Video)
### Cancer Data

```python
from sklearn.model_selection import GridSearchCV

bc_data = datasets.load_breast_cancer()

bc_X_train, bc_X_test, bc_y_train, bc_y_test = train_test_split(
        bc_data['data'], bc_data['target'], test_size=0.2, random_state=42)

tree_param_grid = {'max_depth': [5, 6, 7, 8, 9, 10, 15, 20],
                   'min_samples_leaf': [3, 5, 10, 20],}

bc_tree = DecisionTreeClassifier()

tree_cv = GridSearchCV(bc_tree, tree_param_grid, cv = 5).fit(bc_X_train, bc_y_train)

bc_pred = tree_cv.predict(bc_X_test)

bc_acc = accuracy_score(bc_y_test, bc_pred)

print("Tuned Decision Tree Parameters: " + str(tree_cv.best_params_))
print("Tuned Decision Tree Accuracy: " + str(bc_acc))
```

## Exercise 6: GridSearchCV
### Voting Data

```python
df = pd.read_csv('../datasets/house-votes-84.csv',
                header=None, names = ['infants', 'water', 'budget', 'physician', 'salvador', 'religious',
                                     'satellite', 'aid', 'missile', 'immigration', 'synfuels', 'education',
                                     'superfund', 'crime', 'duty_free_exports', 'eaa_rsa'])

df = df.reset_index()
df.rename(columns = {'index': 'party'}, inplace = True)

df[df == 'y'] = 1
df[df == 'n'] = 0
df[df == '?'] = np.nan
df.iloc[:, 1:] = df.iloc[:, 1:].apply(lambda x: x.fillna(x.mean()))

y = df['party']
X = df.drop('party', axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state=42)

clf = DecisionTreeClassifier()

voting_cv = GridSearchCV(clf, tree_param_grid, cv = 5).fit(X_train, y_train)


y_pred = voting_cv.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print("Tuned Decision Tree Parameters: " + str(voting_cv.best_params_))
print("Tuned Decision Tree Accuracy: " + str(acc))
```

## Residuals

#### Boston

```python
from sklearn import linear_model

boston = datasets.load_boston()

y = boston['target']
X = boston['data']

yb = y.reshape(-1, 1)
Xb = X

# Create linear regression object
regr_all = linear_model.LinearRegression()
# Train the model using the training sets
regr_all.fit( Xb, yb)
mod = regr_all.predict(Xb)

#plot residuals
_ = plt.plot(range(len(Xb)), mod - yb, marker='o', linestyle='None')
```

```python
from sklearn.linear_model import Lasso
las = Lasso(alpha=0.1).fit(Xb, yb)
model = las.predict(Xb)
plt.plot(model-y, marker = 'o', linestyle='None')
plt.figure()
plt.plot(y, marker = 'o', linestyle='None')
plt.plot(model)
plt.show()
```

#### Gapminder Data

```python
df = pd.read_csv('../datasets/gm_2008.csv')
df.drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).head()
Xs = df.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).values
y = df.dropna(how='any').child_mortality.values
y = np.asarray(df.dropna(how='any').child_mortality)
names = df.dropna(how='any').drop(['Country', 'Year', 'Region', 'ID', 'child_mortality'], axis=1).columns

from sklearn.preprocessing import scale
X = scale(Xs)

las = Lasso(alpha=1).fit(X, y)
model = las.predict(X)
plt.plot(model-y, marker = 'o', linestyle='None')
plt.figure()
plt.plot(y, marker = 'o', linestyle='None')
plt.plot(model)
plt.show()
```
