```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
from sklearn import datasets

data = datasets.load_breast_cancer()
```

# ch3_1

## Slide 4

```python
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=0.2, random_state=42)

knn = KNeighborsClassifier(n_neighbors=8)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# ch3_2

## Slide 1

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=0.2, random_state=42)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)

print(confusion_matrix(y_test, y_pred))
#Nprint(classification_report(y_test, y_pred))
```

## Slides 2&3

```python
# Need to use different dataset here: performs too well for instruction!

from sklearn.metrics import roc_curve

y_pred_prob = logreg.predict_proba(X_test)[:,1]

fpr, tpr, thresholds = roc_curve(y_test, y_pred_prob)

plt.figure(figsize=(6, 6), dpi=None)
plt.plot([0, 1], [0, 1], 'k--')
plt.plot(fpr, tpr, label='Logistic Regression')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Logistic Regression ROC Curve')
plt.legend(loc='best')
plt.show()
```

# Slide 5

```python
from sklearn.metrics import roc_auc_score

logreg = LogisticRegression()

X_train, X_test, y_train, y_test = train_test_split(data['data'], data['target'], test_size=0.4, random_state=42)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred_prob = logreg.predict_proba(X_test)[:,1]

print("Area under the Curve: " + str(roc_auc_score(y_test, y_pred_prob)))
```

```python
from sklearn import linear_model

boston = datasets.load_boston()

y = boston['target'].reshape(-1,1)
X = boston['data']

regr = linear_model.LinearRegression()
regr.fit(X, y)
mod = regr.predict(X)

# Plot residuals
_ = plt.plot(range(len(X)), mod - y, marker='o', linestyle='None')
```

# ch3_4

## Slide 3

```python
from sklearn.model_selection import GridSearchCV

breast_cancer = datasets.load_breast_cancer()
X = breast_cancer['data']
y = breast_cancer['target']

# Setup the hyperparameter grid
neighbors = np.arange(1, 50)
param_grid = {'n_neighbors': neighbors}

# Instantiate a k-NN classifier
knn = KNeighborsClassifier()

# Tune the hyperparameter k-NN using GridSearchCV
knn_cv = GridSearchCV(knn, param_grid, cv = 5).fit(X, y)

# Print the tuned parameters and score
print("Tuned k-NN Number of Neighbors: " + str(knn_cv.best_params_))
print("Best score is " + str(knn_cv.best_score_))
```
